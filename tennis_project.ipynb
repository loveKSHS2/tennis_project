# pipeline 만들기

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler())
])
tennis_num_tr = num_pipeline.fit_transform(tennis_num)
tennis_num_tr
```

# train set, test set 분리, rank column drop

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(true_copy_tennis_NoNan,test_size=0.2,random_state=42)
#test_set.head()

tennis_data = train_set.drop(['rank'],axis=1)
tennis_labels = train_set['rank'].copy()
print(tennis_data.info())
```

# num, category attribute 분류, pipeline에 넣어서 데이터 전처리

```python
from sklearn.compose import ColumnTransformer


num_attribs = ['draw_size','tourney_date','match_num','ht','age','rank_points','best_of','minutes','ace'
               ,'df','svpt','1stIn','1stWon','2ndWon','SvGms','bpSaved','bpFaced','win']
cat_attribs = ['surface','tourney_level','hand','round','score','ioc']


full_pipeline = ColumnTransformer([
    ("num",num_pipeline,num_attribs),
    ("cat",OneHotEncoder(),cat_attribs),
])
#tennis_data.head()
tennis_prepared = full_pipeline.fit_transform(tennis_data)
print(tennis_prepared)
print(tennis_prepared.shape)
```

# 모델 훈련

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(tennis_prepared,tennis_labels)
```

```python
from sklearn.metrics import mean_squared_error

tennis_predictions = lin_reg.predict(tennis_prepared)
lin_mse = mean_squared_error(tennis_labels,tennis_predictions)
lin_rmse = np.sqrt(lin_mse)
print(lin_rmse)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(tennis_prepared,tennis_labels)
```

```python
tennis_predictions = tree_reg.predict(tennis_prepared)
tree_mse = mean_squared_error(tennis_labels,tennis_predictions)
tree_rmse = np.sqrt(tree_mse)
print(tree_rmse)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg,tennis_prepared,tennis_labels,scoring="neg_mean_squared_error",cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("점수:",scores)
    print("평균:",scores.mean())
    print("표준편차:",scores.std())

display_scores(tree_rmse_scores)
```

```python
#모델 변경
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=100,random_state=42)
forest_reg.fit(tennis_prepared,tennis_labels)
```

```python
from sklearn.model_selection import cross_val_score
forest_scores = cross_val_score(forest_reg,tennis_prepared,tennis_labels,scoring="neg_mean_squared_error",cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```
